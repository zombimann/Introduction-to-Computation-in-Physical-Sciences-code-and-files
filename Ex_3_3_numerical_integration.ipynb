```python
import numpy as np
from scipy.integrate import quad
```

## Part a

### Straight Numerical Integration

```python
f = lambda x: np.sin(x)**2/x**2
quad(f, 0, np.inf)[0]
```

### Piecewise Numerical Integration - m = 4

```python
a = 100
m = 4
sol = 0
for n in np.arange(0, m):
    sol = sol + list(quad(f, n*a, n*a+a))[0]
sol
```
