```python
import ipywidgets as wgt
from IPython.display import display
```

## Button

```python
b = wgt.Button(description='a button')
output = wgt.Output()

def hello(dummy):
    with output:
        print("My name is Mugambi Ndwiga")
b.on_click(hello)
display(b, output)
```

## Menu

```python
menu = wgt.Select(options=['a', 'b', 'c'], description='select menu')
display(menu) # Dsiplay the menu

b = wgt.Button(description='Print Selected Item')
output = wgt.Output()

def hello(dummy):
    with output:
        print(menu.value) # On button click, print out the value selected on menu
b.on_click(hello)
display(b, output)
```

## Radio buttons Menu

```python
radio = wgt.RadioButtons(options=['a', 'b', 'c'], description='select item')
display(radio) # Dsiplay the menu

b = wgt.Button(description='Print Selected Item')
output = wgt.Output()

def hello(dummy):
    with output:
        print(radio.value) # On button click, print out the value selected on menu
b.on_click(hello)
display(b, output)
```

## Slider

```python
slider = wgt.SelectionSlider(
    options=['Chill', 'Ticking', 'Pissed'],  # List of selectable values
    value='Ticking',                     # Default selected value
    description='Level:',               # Label shown before the slider
    continuous_update=True,             # Update output while sliding
    orientation='horizontal',           # Slider direction
    readout=True                         # Show the selected value
)
display(slider) # Dsiplay the menu

b = wgt.Button(description='Print Selected Item')
output = wgt.Output()

def hello(dummy):
    with output:
        print(slider.value) # On button click, print out the value selected on menu
b.on_click(hello)
display(b, output)
```

## Check

```python
chk = wgt.Checkbox(description='yes')
display(chk) # Display the menu

b = wgt.Button(description='Wanna get punched?')
output = wgt.Output()

def hello(dummy):
    with output:
        print(chk.description) # On button click, print out the value selected on menu
b.on_click(hello)
display(b, output)
```

## Interactive Plots

```python
import matplotlib.pyplot as plt
import numpy as np
def plotf(k, func, refresh):
    if (func == 'sin'):
        f = np.sin(k*np.pi*x)
    elif (func == 'cosh'):
        f = np.cosh(k*np.pi*x)
    else:
        f = np.cos(k*np.pi*x)
    if refresh:
        plt.plot(x, f)
x = np.linspace(-1,1,101)
wgt.interactive(plotf, k=(0,2,.1), func=['sin', 'cos', 'cosh'], refresh=(True))
```
